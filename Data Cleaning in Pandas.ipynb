```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\sylva\OneDrive\Desktop\Portfolio Project\xlsx files\Customer Call List.xlsx")
df
```

```python
# first i will like to drop duplicate values and save it as our new df

df=df.drop_duplicates()
df
```

```python
# secondly lets delete columns that we do not need

df=df.drop(columns='Not_Useful_Column')
df
```

```python
# we can use strip to clear all the unwanted characteristics at once.
# you specify the column you want to work on so that the strip function will not affect other columns or entire table

df['Last_Name']=df['Last_Name'].str.strip('123./_')
df
```

```python
# We can remove these unwanted characteristics individually though not the best approach
# We will use lstrip or rstrip(this will strip the column of all unwanted characters either infront and behind the strings)


df['Last_Name']=df['Last_Name'].str.lstrip('/')
df['Last_Name']=df['Last_Name'].str.lstrip('...')
df['Last_Name']=df['Last_Name'].str.rstrip('_')
df
```

```python
df
```

```python
# we ae going to work to standardise the format for the phone number column
# first we remove anythung that is not a number from that column

df['Phone_Number']=df['Phone_Number'].str.replace('[^a-za-z0-9]','')
df
```

```python
# now we need to set all the values under phone number to string before applying the formating that we want

df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']
```

```python
# apply the formating to the phone numbers above

df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-'+ x[3:6]+'-'+ x[6:10])
df
```

```python
# now lets use replace to get rid of the letters in the numbers column

df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']=df['Phone_Number'].str.replace('a--','')
df
```

```python
df
```

```python
# We are going to work on the address column by splitiing it into street address, state and zipcode
# we use the split function and specify what is currently separating the address

df[['Street_address','State','Zipcode']]=df['Address'].str.split(',',2, expand =True)
df
```

```python
# We will work on paying customer and change Yes to Y and No to N

df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df
```

```python
# lets work on do not contact column and change Yes to Y and No to N

df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
df
```

```python
# now we are going to replace all the N/a and NaN in the entire data

df.replace('N/a','')
```

```python
#we can use the fillna function to remove all cells without values and replace with blank
df=df.fillna('')
df
```

```python
# the task was to give the call center a list of customers that they can call
# we will have to go through the do not contact column and drop any row that has Y in this column

for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x, inplace=True)
df
```

```python
# lets also drop the blanks in the phone number column because the call center cannot call a blank number

for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x, inplace=True)
df  

# we could have also done the above by df=df.dropna(subset='Phone_Number'), inplace=True
```

```python
# We can now reset the index to complete our data cleaning to hand over a clean data to the call center

df=df.reset_index(drop=True)
df
```
