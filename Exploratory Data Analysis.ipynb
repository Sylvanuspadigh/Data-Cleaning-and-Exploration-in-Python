```python
# we can group the data by continent and mean

df.groupby('Continent').mean()
```

```python
# we can sort the above by average 2022 population in descending order

df.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)
```

```python
# we can check the details of what any continent contains

df[df['Continent'].str.contains('Oceania')]
```

```python
# lets assign to a new dataframe df2 for visualization

df2=df.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)
```

```python
# we can transpose the data so that the continents become the headers

df3=df2.transpose()
df3
```
